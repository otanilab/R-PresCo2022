```python
from diffusers import StableDiffusionPipeline
```

```python
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", use_auth_token=TOKEN)
pipe.to("cuda")
```

```python
pipe.safety_checker = lambda images, **kwargs: (images, False)
```

```python
import datetime
from PIL import Image
```

```python
from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/MyDrive
```

```python
prompt = "'realistic photo of club of university', 'live performance', 'participation', 'tastes', 'people', 'friend', 8 k"
for i in range(1000):
  image = pipe(prompt, height=256, width= 384, num_inference_steps=50)
  dt = datetime.datetime.now()
  dt = dt.isoformat()
  image["sample"][0].save(f'/content/drive/MyDrive/img/{dt}.png', quality=100)
  display(image["sample"][0])
```
